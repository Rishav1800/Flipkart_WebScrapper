## Importing Data to CSV

```python
import csv
import requests
from bs4 import BeautifulSoup

# URL of the page 
URL = "https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3Drealme"

# Set the headers to simulate a request from a web browser
HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}

# Send a GET request to the URL with the headers
page = requests.get(URL, headers=HEADERS)

# Create a BeautifulSoup object by parsing the HTML content of the response
soup = BeautifulSoup(page.content, "html.parser")

# Finding the required data
product_title_elems = soup.find_all("div", attrs={'class': '_4rR01T'})
product_price_elems = soup.find_all("div", attrs={'class': '_30jeq3 _1_WHN1'})

# Extracting the text from each element
product_titles = [elem.text for elem in product_title_elems]
product_prices = [elem.text for elem in product_price_elems]

# Writing the data to a CSV file
with open('products.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Product Name', 'Product Price'])
    for title, price in zip(product_titles, product_prices):
        writer.writerow([title, price])

print('Data written to CSV file.')
```
